```python
# Assuming all_df is already defined and contains the 'selected_transition_index' column

# Define a function to count occurrences of the word 'array'
def count_array_occurrences(text):
    return text.count('array')

# Apply the function to the 'selected_transition_index' column and store the result in 'no_transition'
all_df['no_transition'] = all_df['selected_transition_index'].apply(count_array_occurrences)


# Display the first few rows of the DataFrame to verify the changes
all_df.head(20)
```

```python
# replace \n in boxes to comma 
all_df['boxes'] = all_df['boxes'].str.replace('\n', ',')

```

```python
import ast

# Assuming all_df is already defined and contains the specified columns

# Select columns
selected_df = all_df[['peptide_id', 'quantification_scores', 'light_area',
       'light_background', 'heavy_area', 'heavy_background', 'File_Name','no_transition', 'boxes' ]]
col_names = selected_df.columns

# Rename selected column names to alphabet A, B, C, D, E, F, G
selected_df.columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

# Display the first few rows of the DataFrame before extraction
print("DataFrame before extraction:")
print(selected_df.head(20))

# Convert columns B, C, D, E, and F to lists of numbers if they are not already
for col in ['B', 'C', 'D', 'E', 'F']:
    selected_df[col] = selected_df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
    
# convert I to list of string
selected_df['I'] = selected_df['I'].apply(lambda x: x.split(','))

# Display the first few rows of the DataFrame after conversion
print("DataFrame after conversion to lists:")
print(selected_df.head())
```

```python

# Explode the lists in columns B, C, D, E, and F
exploded_df = selected_df.explode(['B', 'C', 'D', 'E', 'F', 'I'])



# I format

# In I column, replace anything in the string before the first number with '['
exploded_df['I'] = exploded_df['I'].apply(lambda x: '[' + x.split('[')[-1])
# In I column replace anything in the string after the last number with ']'
exploded_df['I'] = exploded_df['I'].apply(lambda x: x.split(']')[0] + ']')
# In I column, replace white space with comma 
exploded_df['I'] = exploded_df['I'].apply(lambda x: x.replace(' ', ','))
# repalce [, wuth [
exploded_df['I'] = exploded_df['I'].apply(lambda x: x.replace('[,', '['))
# replace ,] with ]
exploded_df['I'] = exploded_df['I'].apply(lambda x: x.replace(',]', ']'))


# For column I the string should start with '[' then some number then comma then some number then ']'
# exploded_df['I'] = exploded_df['I'].apply(lambda x: '[' + x + ']')


# Rename the columns back to their original names
exploded_df.columns = col_names

# extract first float number from boxes column to boxes_start column
exploded_df['boxes_start'] = exploded_df['boxes'].apply(lambda x: x.split(',')[0])
# extract second float number from boxes column to boxes_end column
exploded_df['boxes_end'] = exploded_df['boxes'].apply(lambda x: x.split(',')[1])
# extract only number in boxes_start and boxes_end column 
exploded_df['boxes_start'] = exploded_df['boxes_start'].apply(lambda x: x.split('[')[-1])
# remove ] from boxes_end column
exploded_df['boxes_end'] = exploded_df['boxes_end'].apply(lambda x: x.split(']')[0])

# Display the first few rows of the exploded DataFrame
print("Exploded DataFrame:")
exploded_df.head(20)

# save to csv
exploded_df.to_csv('/home/thanadol/Documents/GitHub/deep-mrm-KTH/KTH_SRM/test/exploded.csv', index=False)
```
